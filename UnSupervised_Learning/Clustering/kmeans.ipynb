K-means

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("C:\\Users\\mehna\\OneDrive\\Desktop\\Seminar\\UnSupervised_Learning\\Clustering\\Iris.csv")
print(dataset.head(3))
print(dataset.info())
```

```python
plt.figure(figsize=(1, 1))
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(2,21):
    km=KMeans(n_clusters=i, init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
#lets see the elbow graph

# Set figure size before plotting
plt.figure(figsize=(10, 5))
# Plot the data
plt.plot([i for i in range(2, 21)], wcss, marker="o")
plt.xlabel("No. of Clusters")
plt.xticks([i for i in range(2, 21)])
plt.ylabel("WCSS")

# Add grid to the plot
plt.grid(axis="x")

# Show the plot
plt.show()
```

```python
#Lets apply K-means

km=KMeans(n_clusters=3)
dataset["Predict"]=km.fit_predict(dataset)
dataset.to_csv("FinalIris.csv", index=False)
```

```python
print(dataset.head(10))

sns.pairplot(data=dataset, hue="Predict")
plt.show()
```
